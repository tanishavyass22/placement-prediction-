# Feature Matrix In X and Response(Target) In Vector y

```python
X = data.drop('status',axis=1)     #contains all the independent variables
y= data['status']                  #contain all the dependent variables
```

```python
X
```

```python
y
```

# Building the model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
lr = LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
lr.fit(X_train,y_train)
y_pred1 = lr.predict(X_test)

ACC1=accuracy_score(y_test,y_pred1)
print(ACC1)

print(confusion_matrix(y_test,y_pred1))

PCC1=precision_score(y_test,y_pred1)
print(PCC1)

```

```python
knn.fit(X_train,y_train)
y_pred1 = knn.predict(X_test)

ACC2=accuracy_score(y_test,y_pred1)
print(ACC2)

print(confusion_matrix(y_test,y_pred1))

PCC2=precision_score(y_test,y_pred1)
print(PCC2)

```

```python
dt.fit(X_train,y_train)
y_pred1 = dt.predict(X_test)

ACC3=accuracy_score(y_test,y_pred1)
print(ACC3)

print(confusion_matrix(y_test,y_pred1))

PCC3=precision_score(y_test,y_pred1)
print(PCC3)
```

```python
final_data = pd.DataFrame({'Models':['LR','KNN','DT'],
            'Accuracy':[ACC1*100,
                  ACC2*100,
                  ACC3*100],
                'Precision': [PCC1*100,
                       PCC2*100,
                       PCC3*100]})
```
